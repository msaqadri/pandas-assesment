```python
speed_mean = data["Speed"].mean()

def set_speed(val):
    if val < speed_mean:
        return "Speed Low"
    elif val == speed_mean:
        return "Speed Neutral"
    else:
        return "Speed High"
    
```

```python
data["Speed_high_low"] = data["Speed"].apply(set_speed)
data
```

```python
hp_mean = data["HP"].mean()

def set_hp(val):
    if val < hp_mean:
        return "HP Low"
    elif val == hp_mean:
        return "HP Neutral"
    else:
        return "HP High"
```

```python
data["HP_high_low"] = data["HP"].apply(set_hp)
data
```
